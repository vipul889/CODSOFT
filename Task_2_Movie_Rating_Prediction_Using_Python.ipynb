```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Collection And Preprocessing

```python
# collecting data from csv file to pandas
movie_info=pd.read_csv("/content/IMDb Movies India.csv",encoding='latin1')
```

```python
# printing 20 rows from dataset
movie_info.head(20)
```

```python
# printing no of rows and column from the dataset
movie_info.shape
```

```python
# getting the column information from given dataset
movie_info.info()
```

```python
# checking if the given dataset contains any missing values
movie_info.isnull().sum()
```

```python
# There are so many null values so lets remove the rows that contain null val
movie_info.dropna(subset=["Rating"],inplace=True)
```

```python
# rechecking presence of null values
movie_info.isnull().sum()
```

```python
# Removing null values by dropping null values
movie_info.dropna(subset=['Actor 1','Actor 2','Actor 3','Director','Genre'],inplace=True)
```

```python
movie_info['Duration']=movie_info['Duration'].fillna(0).astype(int)
```

```python
# Rechecking for null values again
movie_info.isnull().sum()
```

Data Formatting

```python
movie_info['Votes']=movie_info['Votes'].str.replace(',','').astype(int)
```

```python
movie_info['Year']=movie_info['Year'].str.strip('()').astype(int)
```

```python
movie_info['Duration'] = movie_info['Duration'].str.strip(' min')
```

```python
movie_info.head()
```

```python
movie_info.loc[movie_info['Duration'] == 0, 'Duration'] = movie_info['Duration'].median()
```

```python
movie_info.isnull().sum()
```

```python
movie_info.info()
```

Sorting in descending Order according to rating

```python
#using label based indexing to select rows and column in DataFrame and sorting in descending order
top_cinema=movie_info.loc[movie_info['Rating'].sort_values(ascending=False)[:10].index]
top_cinema
```

Data Visualisation

```python
sns.histplot(data=top_cinema,x='Year',hue='Rating',multiple='stack')
plt.title('Distribution of Top 10 Movies',fontsize=16)
plt.xlabel('Year',fontsize=14)
plt.ylabel('Count',fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# counting the type of genre (i.e drama=count etc)
genre_total=movie_info['Genre'].value_counts().reset_index()
genre_total.columns=['Genre','Count']
top_n_genres=genre_total.head(8)
print(top_n_genres)
```

```python
plt.figure(figsize=(4,4))
plt.pie(top_n_genres['Count'],labels=top_n_genres['Genre'],autopct='%1.1f%%',startangle=100,colors=sns.color_palette('pastel'))
plt.title('Movie Genre Distribution',fontsize=16)
plt.axis('equal')
plt.show()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
L=labelencoder=LabelEncoder()
trans_data=movie_info.drop(['Name'],axis=1)
trans_data['Director']=labelencoder.fit_transform(movie_info['Director'])
trans_data['Actor 1']=labelencoder.fit_transform(movie_info['Actor 1'])
trans_data['Actor 2']=labelencoder.fit_transform(movie_info['Actor 2'])
trans_data['Actor 3']=labelencoder.fit_transform(movie_info['Actor 3'])
trans_data['Genre']=labelencoder.fit_transform(movie_info['Genre'].apply(lambda x: ','.join(x)))
trans_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
sc_data=scaler.fit_transform(trans_data)
sc_df=pd.DataFrame(sc_data,columns=trans_data.columns)
sc_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
X=trans_data.drop(['Rating'],axis=1)
```

```python
y=trans_data['Rating']
```

```python
X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
X_test=np.array(x_test)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
print('R2 score:',r2_score(y_test,y_pred))
print('MSE:',mean_squared_error(y_test,y_pred))
print('MAE:',mean_absolute_error(y_test,y_pred))
```
